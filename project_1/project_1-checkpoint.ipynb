# Bathroom values cleaning

```python
def bathroom_calculator(x):
    try:
        int(x.split(' ')[0])
        return int(x.split(' ')[0])
    except:
        return 1

df1['bathroom'] = df1.bathroom.apply(lambda p: bathroom_calculator(p))
```

# Price cleaning

```python
df1.price.unique()
```

```python
def price_cleaner(x):
    try:
        a,b = x.split(",")
        c = a+b
        d = int(c)
        return d
    except:
        return x
```

```python
df1['price'] = df1.price.apply(lambda p: price_cleaner(p))
df1
```

```python
df1.price.unique()
```

```python
def price_converter(x):
    z = float(x)
    if z<10:
        try:
            y = float(z)
            return y*100000
        except:
            return z
    else:
        return z
df1['price'] = df1['price'].apply(lambda p: price_converter(p))
df1['price'].unique()
```

# One Hot Encoding(Locality,furnish_type,property_type)

```python
df1.head()
```

```python
dummies1 = pd.get_dummies(df1['locality'])
dummies1.head()
```

```python
dummies2 = pd.get_dummies(df1['property_type'])
dummies2.head()
```

```python
dummies3 = pd.get_dummies(df1['furnish_type'])
dummies3.head()
```

```python
df2 = pd.concat([df1,dummies1,dummies2,dummies3],axis=1)
df2.head()
```

```python
df2.drop(['property_type','locality','furnish_type'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
df2.shape
```

# Model Building

```python
x=df2.drop('price', axis=1)
```

```python
y=df2.price
```

```python
y = y.apply(lambda p: int(p))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
m1 = Lasso()
m1.fit(x_train, y_train)
m1.predict(x_test)
```

```python
m1.score(x_test,y_test)
```

```python
m2 = DecisionTreeRegressor()
m2.fit(x_train, y_train)
m2.score(x_test,y_test)
```
